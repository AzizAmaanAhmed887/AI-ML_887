```python
# Stack Plot

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 4, 3, 6]
y2 = [2, 3, 5, 1, 2]
y3 = [0, 1, 2, 4, 3]

# Basic stacked area plot
plt.stackplot(x, y1, y2, y3, labels=['Layer 1', 'Layer 2', 'Layer 3'])

# Adding labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Basic Stacked Area Plot')
```

```python
# subplots

import numpy as np

x = [1, 2, 3, 4, 5]

y1 = [np.sqrt(i) for i in x]
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]
y4 = [i*2 for i in x]
# ploting these 4 plot inside a single figure using subplots

plt.subplot(2, 2, 1)
plt.plot(x, y1, label='sqrt(x)')
plt.title('Square Root')

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='x^2')
plt.title('Square')

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='x^3')
plt.title('Cube')

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='2x')
plt.title('Linear')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].plot(x, y1, label="sqrt(x)")
ax[0, 0].set_title("Plot1 - Square Root")

ax[0, 1].plot(x, y2, label="x^2")
ax[0, 1].set_title("Plot2 - Square")

ax[1, 0].plot(x, y3, label="x^3")
ax[1, 0].set_title("Plot3 - Cube")

ax[1, 1].plot(x, y4, label="2x")
ax[1, 1].set_title("Plot4 - Linear")

fig.suptitle("Subplots Example", fontsize=16)

fig.tight_layout()
plt.show()
```

```python
# Weekly temperature data for two cities
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]
temp = [
    [30, 32, 31, 29, 28],  # New York
    [22, 24, 23, 21, 20],  # London
    [35, 37, 36, 34, 33],  # Delhi
    [25, 27, 26, 24, 23],  # Tokyo
]

# Create a figure and axis
fig, axes = plt.subplots(2, 2)

city_no = 0
for i in range(2):
    for j in range(2):
        axes[i][j].plot(days, temp[city_no], marker='o')
        axes[i][j].set_title(cities[city_no])
        axes[i][j].grid()
        city_no += 1

fig.suptitle("Weekly Temperature Trends in Different Cities", fontsize=16)
plt.tight_layout()
        
```
