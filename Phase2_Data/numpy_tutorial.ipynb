```python
# Vectorization

arr = np.array([1, 2, 3, 4, 5])

# Without vectorization
final = []
for i in arr:
    final.append(i * 2)
print("Without vectorization")
print(final)

# With vectorization
result = arr * 2
print("With vectorization")
print(result)
```

```python
# Broadcasting
# a = np.array([1, 2, 3])
# b = np.array([5, 6, 7], [8, 9, 10])  # -> [10 ,10, 10] broadcasted

# print(a + b)

a1 = np.array([[1, 2, 3], [4, 5, 6]])
b1 = np.array([7, 8, 9])

print(a1.shape)  # (2, 3)
print(b1.shape)  # (3, )

print(a1 + b1)
```

```python
arr = np.array([
    [1, 2],
    [3, 4]
])

mean = np.mean(arr)
std_dev = np.std(arr)

normalized_arr = (arr - mean) / std_dev

# print(normalized_arr)
# print(mean)
# print(std_dev)

print(np.sqrt(5 / 4))
```

```python
# Mathematical functions - Aggregation fnx
arr = np.array([1, 2, 3, 4, 5])

print("Sum = ", arr.sum())
print("Product = ", arr.prod())
print("minimum = ", arr.min())
print("maximum = ", arr.max())
print("Minimum index, ", arr.argmin())
print("Maximum index, ", arr.argmax())
print("Variance = ", arr.var())
```

```python
# power fnx
arr = np.array([1, 2, 3, 4, 5])

print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr, 3))
```

```python
# log and exponential
arr = np.array([1, 2, 3, 4, 5])

# log methods
print(np.log(arr))
print(np.log2(arr))
print(np.log10(arr))

# exponent methods
print(np.exp(arr))
```

```python
# rounding condition
arr = np.array([1, 2, 3, 5, 5])

print(np.round(3.14))
print(np.ceil(3.14))
print(np.floor(3.14))
print(np.trunc(56.8895))
print(np.unique(arr))

arr2 = np.array([-1, 2, -3, 4, -5, 6, -7])
print(np.abs(arr2))
```
