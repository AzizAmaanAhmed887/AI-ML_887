```python
# lecture 2: Line Plot

# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

# plot 
plt.plot(oscar_years, oscar_revenue, marker='o', color='red')
plt.title("Oscar Winner Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
plt.grid()
plt.show()
```

```python
# lecture 3: Multiple Line Plots

# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

# Oscar Movies Revenue vs Non-Oscar Movies Revenue
plt.plot(oscar_years, oscar_revenue, "o--r", label='Oscar Winners')
plt.plot(oscar_years, non_oscar_revenue, marker='o', color='blue', label='Non-Oscar Winners')

# plt.title("Oscar vs Non-Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")
plt.legend()
plt.grid()

plt.show()
```

```python
# Bar chart

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo",
]
years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(oscar_movies, revenue, color='green')
plt.title("Oscar Winner Movies Revenue")
plt.xlabel("Movie")
plt.ylabel("Revenue (in $M)")
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent overlap

for i in range(len(oscar_movies)):
    plt.text(i, revenue[i], str(revenue[i]), ha='center')

plt.ylim(0, max(revenue)+120)
plt.show()
```

```python
# comparing data with bar charts
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo",
]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
years = [2008, 2009, 2010, 2011, 2012]

# Fix Graph
import numpy as np
x = np.arange(len(years))  # the label locations -> [0, 1, 2, 3, 4]
width = 0.4

plt.bar(
    x - width / 2, oscar_revenue, width=width, color="red", label="Oscar Movies Revenue"
)
plt.bar(
    x + width / 2,
    non_oscar_revenue,
    width=width,
    color="blue",
    label="Non-Oscar Movies Revenue",
)

plt.title("Oscar vs Non-Oscar Movies Revenue")
plt.xlabel("Movies")
plt.ylabel("Revenue (in $M)")

plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent overlap

plt.legend()
plt.show()
```

```python
# horizontal bar chart

oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo",
]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.barh(oscar_movies, oscar_revenue, color='purple', label = "Oscar Movies Revenue", alpha=0.7)
plt.title("Oscar Winner Movies Revenue")

plt.xlabel("Revenue (in $M)")
plt.ylabel("Movie")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.grid()
plt.legend()
plt.show()
```

```python
# Scatter Plot

people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

# if bp > 130 -> High BP(color red) else -> Normal BP(color green)
colors = ['red' if bp >= 130 else 'green' for bp in blood_pressure]

plt.scatter(age, blood_pressure,marker='^', s=blood_pressure,cmap='OrRd', c=blood_pressure)
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.grid()

plt.colorbar(label='Blood Pressure (in mmHg)')

# Providing Annotations to all points
for i in range(len(people)):
    plt.annotate(
        people[i],
        xy=(age[i], blood_pressure[i]),  # Point to annotate
        xytext=(age[i]+3, blood_pressure[i]+3),  # Position of the annotation text
    )

plt.xlim(min(age), max(age)+10)
plt.ylim(min(blood_pressure), max(blood_pressure)+10)

plt.show()
```

```python
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (°C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (°C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

plt.scatter(winter_temp, winter_humidity, marker='o', color='blue', label='Winter')
plt.scatter(summer_temp, summer_humidity, marker='^', color='orange', label='Summer')
plt.title("Temperature vs Humidity in Different Cities")

plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")

plt.grid()
plt.legend()

plt.show()
```

```python
# Pie Charts

labels = ["Python", "JavaScript", "Java", "C++", "Ruby"]
popularity = [30, 25, 20, 15, 10]

plt.pie(
    popularity,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    explode=[0.1, 0, 0, 0, 0],
    wedgeprops={
        "width": 0.5,
        "edgecolor": "white",
        "linewidth": 2,
    },
    shadow=True,
)
plt.title("Programming Language Popularity")
plt.show()
```
